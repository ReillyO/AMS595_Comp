## 2. Implementing Gradient Descent Algorithm (60 pts)

Given a function $f(x) = \sqrt{x^2+5}$, $x\in \mathbf{R}$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x).

func = lambda x : np.sqrt(x**2 + 5)

# Plot the function f(x), np.linspace might be helpful.
x_vals = np.linspace(-5, 5, 100) 
y_vals = func(x_vals)

ax1 = plt.plot(x_vals, y_vals, "b-")
```

```python
# Define the derivative of the function f(x), for gradient descent
dfunc = lambda x : x/np.sqrt(5+x**2)

# write a function to run the Gradient Descent Algorithm, take input arguments x_0, step_size, and num_iterations; return a list/array of x_k values.

def line_descent(x_0, step_size, num_iterations):
    x = x_0
    x_k = np.array([])
    for i in range(num_iterations):
        steep = dfunc(x)
        x -= steep * step_size
        x_k = np.hstack([x_k, x])
    return x_k

# set up/assign initial values
x_0 = 7 / 5 + 1
NITER = 50

# Run gradient descent by calling the function
x_k = line_descent(x_0, .5, NITER)

# report x_T by printing out the value
print(x_k[len(x_k)-1])

# Plot the f(x_k) values vs. iteration number
plt.plot(x_k)
plt.xlabel("Iteration")
plt.ylabel("Value")
```

```python
# create a list containing step sizes
step_sizes = np.array([0.5, 1, 3, 5])

# initialize a list to store the results from different step sizes
results = np.zeros(50)

# loop through the step sizes and perform GD and record results
for step in step_sizes:
    results = np.vstack([results, line_descent(x_0, step, NITER)])
results = results[1:]


# plot multiple curves in a single graph
for i in range(len(step_sizes)):
    plt.plot(results[i], label=f"Step {step_sizes[i]}")
plt.title("Gradient Descent Training")
plt.xlabel("Iterations")
plt.ylabel("Value Guess")
plt.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import cm

# Generate synthetic data
np.random.seed(27)   # change the seed to be the last two digits of your SB ID.
X = 2 * np.random.rand(100, 2)  # Generate 100 random 2D data points from 0 to 2
y = 4 + 4 * X[:, 0] + 0.5 * X[:, 1] + np.random.randn(100)  # Generate y values with noise

# Add a column of ones to X for the bias terms (intercept) in the normal equation
X = np.hstack([X, np.ones((100, 1))])

# generate meshgrid for plotting prediction
X1, X2 = np.meshgrid(X[np.linspace(0, len(X[:,0])-1, 10, dtype=int), 0], X[np.linspace(0, len(X[:,1])-1, 10, dtype=int), 1])

# Find theta using the normal equation
# normal eq process
XtX = np.dot(np.transpose(X), X)
Xty = np.dot(np.transpose(X), y)
theta = np.linalg.solve(XtX, Xty)

# generate plane coords
predict = np.dot(X, theta)
predict = predict[np.linspace(0, len(predict)-1, 10, dtype=int)]
predict_grid = np.meshgrid(predict, predict)[0]

# Print the intercept and coefficients of the regression plane
print(f"Intercept: {theta[0]}")
print(f"Coefficients: {theta[1:3]}")

# Plot the generated data and the regression plane
# setup
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.view_init(elev=30,azim=75,roll=0)

# plot and label
ax.scatter(X[:,0], X[:,1], y, 'bo', label="Original data points")
ax.plot_surface(X1, X2, predict_grid, cmap=cm.viridis)
ax.set_xlabel("X1")
ax.set_ylabel("X2")
ax.set_zlabel("Vals")
ax.legend()

```

## 4. Logistic Regression with Numpy and Graphing with Matplotlib (20 pts)

In class, we implemented a logistic regression model to perform binary classification. Here, I provided the exact same code that we discussed in class, but the values for learning_rate and num_epochs are modified. I also removed all the comments. You task is to
1. Tune the Learning Rate and Number of Epochs: Your first task is to fine-tune the values for the learning_rate and num_epochs parameters. Your goal is to identify suitable values that will enable us to converge to a set of parameter values $\theta$ closely approximating the optimal parameter values $\theta^*$.
2. Code Explanation: To enhance code comprehension, please augment the code with meaningful comments. These comments should elucidate the purpose and functionality of each code segment, making it easier for readers to understand the logistic regression implementation.

By accomplishing these tasks, we aim to achieve a better understanding of the logistic regression model's behavior and its parameter optimization process.

```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoidal activation function 
def sigmoid(z):
    return 1 / (1 + np.exp(-z))


np.random.seed(595)
X = np.random.rand(2, 100) # random starting values 
w_true = np.array([1.5, -2.5]) # True "template" weight
b_true = 1.0  # True "template" bias
probabilities = sigmoid(np.dot(w_true.T, X) + b_true) # Generate ground truth of activation
Y = (probabilities > 0.5).astype(int) # Y as true/false training array based on activation
X = X + 0.3 * np.random.rand(2, 100) - 0.1 * np.random.rand(2, 100) # Add noise to X in preparation for training

# Divide noisy X and ground truth Y into training and test sets
X_train, X_test = X[:,:80], X[:,80:]
Y_train, Y_test = Y[:80], Y[80:]

# set initial weights/biases and training hyperparams
w = np.zeros(X.shape[0])
b = 0.0
learning_rate = 20
num_epochs = 100000

# iterate through epochs and progressively improve weights/biases
for epoch in range(num_epochs):
    A_train = sigmoid(np.dot(w.T, X_train) + b)

    # set weight "velocity" to normalized fraction of correct values
    dJdw = np.dot(X_train, (A_train - Y_train).T) / len(Y_train)
    # and bias to average difference between ground truth and produced values
    dJdb = np.mean(A_train - Y_train)

    # update weight and bias based on new rates
    w -= learning_rate * dJdw
    b -= learning_rate * dJdb

# generate final predictions and count number of activations for training set
A_train = sigmoid(np.dot(w.T, X_train) + b)
predictions_train = (A_train > 0.5).astype(int)

# generate final predictions and count number of activations for test set
A_test = sigmoid(np.dot(w.T, X_test) + b)
predictions_test = (A_test > 0.5).astype(int)

# calculate accuracy as fraction of true activations produced for test and training sets
train_accuracy = np.mean(predictions_train == Y_train)
test_accuracy = np.mean(predictions_test == Y_test)

# report accuracy
print(f"Training Set Accuracy: {train_accuracy:.4f}")
print(f"Test Set Accuracy: {test_accuracy:.2f}")
```

```python
# Plot true activation and false activation points for TRAINING set
plt.scatter(X[0, :80], X[1,:80], c=Y[:80], cmap=plt.cm.Paired)
plt.xlim(-0.2, 1.2)
plt.ylim(-0.2, 1.2)
ax = plt.gca()

# Generate a boundary line for the logistic regression based on 
# the weights and biases generated in the earlier block from training
xlim = ax.get_xlim()
ylim = ax.get_ylim()
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 50), np.linspace(ylim[0], ylim[1], 50))
Z = np.dot(w.T, np.c_[xx.ravel(), yy.ravel()].T) + b # uses trained weights + biases
Z = sigmoid(Z)
Z = Z.reshape(xx.shape)
plt.contour(xx, yy, Z, colors='k', levels=[0.5], alpha=0.5, linestyles=['--'])

# add labels
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Logistic Regression Decision Boundary on Training Set')
plt.show()
```

```python
# Plot true activation and false activation points for TRAINING set
plt.scatter(X[0, 80:], X[1, 80:], c=Y[80:], cmap=plt.cm.Paired)
plt.xlim(-0.2, 1.2)
plt.ylim(-0.2, 1.2)

# Generate a boundary line for the logistic regression based on 
# the weights and biases generated in the earlier block from training
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 50), np.linspace(ylim[0], ylim[1], 50))
Z = np.dot(w.T, np.c_[xx.ravel(), yy.ravel()].T) + b
Z = sigmoid(Z)
Z = Z.reshape(xx.shape)
plt.contour(xx, yy, Z, colors='k', levels=[0.5], alpha=0.5, linestyles=['--'])

# add labels
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Logistic Regression Decision Boundary on Test Set')
plt.show()
```
